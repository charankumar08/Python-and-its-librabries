## files operation

```python
with open ('example.txt','r') as file:
    content = file.read()
    print(content)
```

```python
with open ('example.txt','r')as file:
    for line in file:
        print(line) ##strip() removes the new line
        
```

```python
## (overwriting)
with open('example.txt','w') as file:
    file.write('hello world!\n') 
    file.write('this is a new line.') 
    
```

```python
##w/o overwriting

with open('example.txt','a') as file:
    file.write("append operation tahing place!")
```

```python
## writing a list of lines to a file
lines=['\nFirst line \n', 'second line \n','third line\n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
#writing to a binary file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
#reading a binary file
with open('example.bin','rb') as file:
    binary_data = file.read()
    print(binary_data)
```

```python
## read the content from  a source data file and write to a copy_example file
# copying a text file
with open('example.txt','r') as source_file:
    content = source_file.read()
with open('copy_example.txt','w') as dest_file:
    dest_file.write(content)
```

```python
## read a text file and count the number of lines,words and characters
def count_file_content(filename):
    with open(filename,'r') as file:
        content = file.read()
        lines = content.splitlines()
        words = content.split()
        characters = len(content)
        print(f'Lines: {len(lines)}')
    print(f'Words: {len(words)}')
    print(f'Characters: {characters}')
    return lines, words, characters

file_path = 'example.txt'
lines, words, characters = count_file_content(file_path)
print(f'Lines: {len(lines)}, Words: {len(words)}, Characters: {characters}')
```
