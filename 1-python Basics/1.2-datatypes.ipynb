### Data types


```python
## integers

a = 10
print(a)
```

```python
## floating point numbers
b = 3.14
print(b)
print(type(b))
```

```python
## string data type
name = "cherry"
print(name)
print(type(name))

```

```python
## boolean data type
is_true = True
print(is_true)
type(is_true)
is_false = False
print(is_false)
type(is_false)
```

```python
a= 20
b= 20

type (a==b)
```

```python
## common errors
result = "hello" + 10
print(result)
```

```python
result = "hello" + str(10)
print(result)
```
