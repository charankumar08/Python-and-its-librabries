```python
## nameing convention
## 1. variable name should start with a letter or underscore
## 2. variable name can contain letters, numbers and underscores
## 3. variable name should not be a keyword
## 4. variable name should be descriptive
## valid variable names
age = 20
height = 5.5
name = "cherry"
is_student = True

## invalid variable names
# 1age = 20
# height@ = 5.5
# name# = "cherry"
# is student = True

## example
first_name = "cherry"
last_name = "charan"
full_name = first_name + " " + last_name
print("full name :" ,full_name)
```

```python
## understanding the variable
## python is a dynamically typed language, which means that we don't need to declare the type of variable. The type of variable is determined at runtime based on the value assigned to it. We can change the value of variable and its type at any time.
age = 20
print(type(age)) ## age is an integer
age = "twenty"
print(type(age)) ## age is a string
```

```python
## type checking and conversion

age = 20
print(type(age))

# type conversion
age_str = str(age)
print(age_str)
print(type(age_str))

```

```python
## dynamic typing
## python allows us to change the type of variable at runtime. This is called dynamic typing. We can assign a value of any type to a variable and it will work without any error.

var = 10
print(var, type(var))

var = "hello"
print(var, type(var))

var = 3.14     
print(var, type(var))
```

```python
## input from user
name = input("enter your name : ")
print("hello", name)
print(name, type(name)) ## input from user is always a string
```
