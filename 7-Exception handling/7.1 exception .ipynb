## Excetion handling

```python
## exception try,except block
try:
    a=b
except:

    print("Variable 'b' is not defined.")
```

```python
try:
    a=b
except NameError as ex:
    print(f"Error: {ex}")
```

```python
try:
    result =1/0
except ZeroDivisionError as ex:
    print(f"Error: {ex}")
    print("Cannot divide by zero.")
```

```python
try :
    result = 1/0
except Exception as ex:
    print(ex)
    print("An error occurred during division.")
except Exception as ex:
    print(f"An error occurred: {ex}")
    print("main exception block caught here.")
```

```python
try:
    num=int(input("Enter a number: "))
    result = 10/num
except ValueError as ex:
    print(f"Invalid input: {ex}")
except ZeroDivisionError as ex:
    print(f"Error: {ex}")
    print("Cannot divide by zero.") 
except Exception as ex:
    print(f"An unexpected error occurred: {ex}")
```

```python
##try,except,else block
try:
    num=int(input("Enter a number: "))
    result = 10/num
except ValueError:
    print("Invalid input. Please enter a valid integer.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
else:
    print(f"The result of 10 divided by {num} is: {result}")
    
```

```python
## try ,excet,else,finally block
try:
    num=int(input("Enter a number: "))
    result = 10/num 
except ValueError:
    print("Invalid input. Please enter a valid integer.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
except Exception as ex:
    print(f"An unexpected error occurred: {ex}")
else:
    print(f"The result of 10 divided by {num} is: {result}")
```
