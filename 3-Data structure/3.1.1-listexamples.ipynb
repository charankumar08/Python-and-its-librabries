```python
## to do list

to_do_list = [ "buy groceries", "clean the house", "pay bills" ]

## adding to task
to_do_list.append("call mom")
to_do_list.append("finish project")

## removing a task
to_do_list.remove("clean the house")

## checking if a task is in the list
if "pay bills" in to_do_list:
    print("Don't forget to pay the bills!")

## printing the to-do list
print("To-Do List:")
for task in to_do_list:
    print(f"- {task}")
        
```

```python
## organizing student grades

grades = [86, 92, 78, 90, 85]

grades.append(88)  # Adding a new grade

avg_grade = sum(grades) / len(grades) 
print(f"Average grade: {avg_grade}")

highest_graqde = max(grades)
lowest_grade = min(grades)
print(f"Highest grade: {highest_graqde}")
print(f"Lowest grade: {lowest_grade}")
```

```python
## managing a inventory

inventory = ["apple", "banana", "orange"]

inventory. append("grape")
inventory.remove("banana")


item ="orange"
if item in inventory:
    print(f"{item} is in stock.")

else:
    print(f"{item} is out of stock.")


print("Current inventory:")
for item in inventory:
    print(f"- {item}")    

```

```python
## collecting user feedback 

feedback = ["Great service!", "Could be better.", "Loved the experience!" ]

feedback.append("Will recommend to others.")
positive_feedback = [f for f in feedback if "Great" in f or "Loved" in f or "recommend" in f]
print("Positive Feedback:")

for f in positive_feedback:
    print(f"- {f}")


```
