## Dictionaries

```python
empty_dict={}
print(empty_dict)
print(type(empty_dict))
```

```python
student = { "name": "cherry", "age": 20, "grade": "A" }
print(student)
print(type(student))
```

```python
## accessing values

student = { "name": "cherry", "age": 20, "grade": "A" }

print(student["name"])  # Output: cherry
print(student["age"])   # Output: 20
print(student["grade"]) # Output: A


## adding and updating key-value pairs
student = { "name": "cherry", "age": 20, "grade": "A" }
student["age"] = 21  # Update existing key
student["major"] = "Computer Science"  # Add new key-value pair
print(student)
# Output: {'name': 'cherry', 'age': 21, 'grade': 'A', 'major': 'Computer Science'}


## removing key-value pairs
student = { "name": "cherry", "age": 20, "grade": "A" }
del student["grade"]  # Remove key-value pair with key "grade"
print(student)  # Output: {'name': 'cherry', 'age': 20}
student.pop("age")  # Remove key-value pair with key "age" and return its value
print(student)  # Output: {'name': 'cherry'}
student.clear()  # Remove all key-value pairs from the dictionary   
print(student)  # Output: {}


## dictionary methods
student = { "name": "cherry", "age": 20, "grade": "A" }
print(student.keys())    # Output: dict_keys(['name', 'age', 'grade'])
print(student.values())  # Output: dict_values(['cherry', 20, 'A'])
print(student.items())   # Output: dict_items([('name', 'cherry'), ('age', 20), ('grade', 'A')])
student["age"] = 21
print(student.get("age"))  # Output: 21
print(student.get("major", "Not Found"))  # Output: Not Found
student["address"] = "123 Main St"
print(student)  # Output: {'name': 'cherry', 'age': 21, 'grade': 'A', 'address': '123 Main St'}
```

```python
## shallow copy 

original_dict = {"name": "cherry", "age": 20, "grade": "A"}
shallow_copy = original_dict.copy()
shallow_copy["age"] = 21
print(original_dict)  # Output: {'name': 'cherry', 'age': 20, 'grade': 'A'}
print(shallow_copy)   # Output: {'name': 'cherry', 'age': 21, 'grade': 'A'}

#deep copy
import copy
original_dict = {"name": "cherry", "age": 20, "grade": "A"}
deep_copy = copy.deepcopy(original_dict)
deep_copy["age"] = 21
print(original_dict)  # Output: {'name': 'cherry', 'age': 20, 'grade': 'A'}
print(deep_copy)      # Output: {'name': 'cherry', 'age': 21, 'grade': 'A'}
```

```python
### iterating over a dictionary
## you can use a for loop to iterate over the keys, values, or key-value pairs of a dictionary. Here are some examples:     

for keys in student.keys():
    print(keys)  # Output: name, age, grade\n   
for values in student.values():
    print(values)  # Output: cherry, 20, A\n

for key, value in student.items():
    print(f"{key}: {value}")  # Output: name: cherry\n age: 20\n grade: A\n

```

```python
## nested dictionaries

students={
    "student1": {"name": "cherry", "age": 20, "grade": "A"},
    "student2": {"name": "kittu", "age": 22, "grade": "B"},
    "student3": {"name": "vinay", "age": 21, "grade": "A"}
}
print(students)
```

```python
## acesssing nested dictionary values

print(students["student1"]["name"])  # Output: cherry
print(students["student2"]["age"])   # Output: 22
print(students["student3"]["grade"]) # Output: A
```

```python
## iterating over nested dictionaries

for student_id, student_info in students.items():
    print(f"{student_id}:")
    for key, value in student_info.items():
        print(f"  {key}: {value}")
```

```python
## dictionary comprehension

squared_numbers = {x: x**2 for x in range(1, 6)}
print(squared_numbers)  # Output: {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```

```python
## conditional dictionary comprehension

even_squares = {x: x**2 for x in range(1, 11) if x % 2 == 0}
print(even_squares)  # Output: {2: 4, 4: 16, 6: 36, 8: 64, 10: 100} 
```

```python
## example

numbers = [1,2,2,3,3,3,4,4,4,4]
frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number] +=1
    else:
        frequency[number] = 1

print(frequency)  # Output: {1: 1, 2: 2, 3: 3, 4: 4}
```

```python
## marge two dictionaries

dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4}

merged_dict = {**dict1, **dict2}
print(merged_dict)  # Output: {'a': 1, 'b': 3, 'c': 4}

```
