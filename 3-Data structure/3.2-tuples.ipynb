## Tuples


```python
lst= list()
print(type(lst))
tpl = tuple()
print(type(tpl))

```

```python
numbers = tuple([1, 2, 3, 4, 5])
print(numbers)  # Output: (1, 2, 3, 4, 5)
```

```python
mixed_tuple = (1, "hello", 3.14, [1, 2, 3], (4, 5))
print(mixed_tuple)  # Output: (1, 'hello', 3.14, [1, 2, 3], (4, 5)) 
```

```python
## accessing tuple elements

my_tuple = (10, 20, 30, 40, 50)
print(my_tuple[0])  # Output: 10
print(my_tuple[2])  # Output: 30
print(my_tuple[-1])  # Output: 50
print(my_tuple[1:4])  # Output: (20, 30, 40)
```

```python
## tuple operations

concatenated_tuple = (1, 2, 3) + (4, 5)
print(concatenated_tuple)  # Output: (1, 2, 3, 4, 5)
```

```python
## immutable nature of tuples

lst = [1,2,3,4,5]
print(lst)

lst[1]="cherry"
print(lst)

```

```python
## tiples methods

print(numbers.count(1))
print(numbers.index(3))
```

```python
## packing and unpacking tuples

packed_tuple = 1, "hello", 3.14
print(packed_tuple)  # Output: (1, 'hello', 3.14)
```

```python
## unbpacking

a,b,c = packed_tuple

print(a)
print(b)
print(c)
```

```python
## unpacking with *

numbers = (1,2,3,4,5,6,7)
first, *middle, last = numbers
print(first)
print(middle)
print(last)
```

```python
## nested tuples

lst = [[1,2,3] ,[4,5,6], [7,8,9]]
tpl = ((1,2,3), (4,5,6), (7,8,9))

lst[0][1] = "cherry"
print(lst)

# tuples are immutable; convert to a mutable structure, modify, then convert back
tpl_list = [list(x) for x in tpl]
tpl_list[0][1] = "cherry"
tpl = tuple(tuple(x) for x in tpl_list)
print(tpl)
```

```python
nested_tuple =  ((1, 2, 3), ("a","b","c"), (True, False, None))
print(nested_tuple[0])
print(nested_tuple[1])
print(nested_tuple[2])
print(nested_tuple[2][2])
```

```python
## iterating through a tuple

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()    
```
