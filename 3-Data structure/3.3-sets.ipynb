## sets

```python
## creating a set

my_set = {1, 2, 3, 4, 5}
print(my_set)  # Output: {1, 2, 3, 4, 5}
print(type(my_set))  # Output: <class 'set'>
```

```python
my_set = set([1, 2, 3, 4, 5]) 
print(my_set)  # Output: {1, 2, 3, 4, 5}
print(type(my_set))  # Output: <class 'set'>
```

```python
my_empty_set = set([1,1,2,3,4,5,5,6,7,])
print(my_empty_set)  # Output: set()
```

```python
## basic set operations

my_set.add(8)
print(my_set)
my_set.add(2)
print(my_set)

my_set.remove(2)
print(my_set)
my_set.discard(10)
print(my_set)
my_set.discard(3)
print(my_set)

```

```python
## pop method

popped_element = my_set.pop()
print(popped_element)  # Output: 1 (or any other element, since sets are unordered)
print(my_set)  # Output: {2, 3, 4, 5, 6, 7, 8} (or the remaining elements after popping)
my_set.clear()
print(my_set)  # Output: set()
```

```python
## set membership

my_set = {1, 2, 3, 4, 5}
print(3 in my_set)  # Output: True
print(6 in my_set)  # Output: False
```

```python
## mathamatical set operations

set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

union_set = set_a.union(set_b)
print(union_set)  # Output: {1, 2, 3, 4, 5, 6, 7, 8}
intersection_set = set_a.intersection(set_b)
print(intersection_set)  # Output: {4, 5}
difference_set = set_a.difference(set_b)
print(difference_set)  # Output: {1, 2, 3}
symmetric_difference_set = set_a.symmetric_difference(set_b)
print(symmetric_difference_set)  # Output: {1, 2, 3, 6, 7, 8}

```

```python
## sets methods

set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

print(set_a.issubset(set_b))  # Output: False
print(set_a.issuperset(set_b))  # Output: False
print(set_a.isdisjoint(set_b))  # Output: False
```

```python
lst = [1, 2, 3, 4, 5, 5, 6, 7]

set(lst)
```

```python
### counting unique elements in a list

text = "in the middle of the night"
words = text.split()

unique_words = set(words)
print(unique_words)  # Output: {'in', 'the', 'middle', 'of', 'night'}
print(len(unique_words))  # Output: 5
```
