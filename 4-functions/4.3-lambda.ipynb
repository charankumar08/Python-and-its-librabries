## lambda

```python
lambda arguments: expression
```

```python
def addition(x,y):
    return x+y

addition(2,3)
```

```python
addition=lambda a,b :a+b
type(addition)

addition(2,3)
```

```python
def even (num):
    if num%2==0:
        return True
    else:
        return False
even(4)


even=lambda num: True if num%2==0 else False
even(5)
```

```python
def addition (x,y,z):
    return x+y+z
addition (12,13,14)
```

```python
addition = lambda x,y,z : x+y+z
addition (12,13,14)
```

```python
## map()

numbers = [1,2,3,4,5]

def square (number):
    return number**2

square(2)
```

```python
list (map(lambda x:x**2,numbers))
```
