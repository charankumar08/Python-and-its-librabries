## map() function

```python
def square(x):
    return x**2
square(2)
```

```python
numbers = [1,2,3,4,5]
list (map(square,numbers))
```

```python
## lambda function woth map 
numbers = [1,2,3,4,5]
list(map(lambda x:x*x,numbers))
```

```python
## mul iterables
numbers1 = [1,2,3,4,5]
numbers2 = [6,7,8,9,10]

added_numbers = list(map(lambda x,y : x+y,numbers1,numbers2))
print(added_numbers)
```

```python
int_numbers = [1,2,3,4,5]
str_numbers = list(map(lambda x:str(x),int_numbers))
print(str_numbers)
```

```python
words = ['python','is','awesome']
upper_words= list(map(str.upper,words))
print(upper_words)
```

```python
def get_name(person):
    return person['name']

people = [{'name':'Alice','age':25},
            {'name':'Bob','age':30},
            {'name':'Charlie','age':35}]

names = list(map(get_name,people))
print(names)
```
